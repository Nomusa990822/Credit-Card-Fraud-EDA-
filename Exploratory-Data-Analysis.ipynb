## 9. Data Cleaning for Analysis

Some columns are identifiers or personal details that are not useful for exploratory fraud pattern analysis in their raw form. These include row identifiers, names, card numbers, street addresses, and transaction identifiers.

Removing them makes the analysis cleaner and more focused.

```python
columns_to_drop = [
    "Unnamed: 0",
    "cc_num",
    "first",
    "last",
    "street",
    "trans_num"
]

df = df.drop(columns=[col for col in columns_to_drop if col in df.columns])

print("Updated shape:", df.shape)
print(df.columns.tolist())
```

### Insight

A total of **6 columns** were removed from the dataset, reducing the feature count from **23 to 17 columns**, while preserving all **1,852,394 transactions**.

The removed columns (`Unnamed: 0`, `cc_num`, `first`, `last`, `street`, `trans_num`) primarily serve as:
- unique identifiers
- personal information
- transaction-specific references

These variables do not contribute meaningful analytical signal for fraud detection and may introduce noise or unnecessary complexity.

The remaining dataset focuses on:
- **transactional features** (e.g., `amt`, `category`)
- **geographic features** (e.g., `state`, `lat`, `long`, `city_pop`)
- **temporal features** (e.g., `trans_date_trans_time`, `unix_time`)
- **target variable** (`is_fraud`)

This refinement ensures that subsequent analysis is centered on variables that are more likely to explain fraud behavior rather than merely identify individual transactions.

From a modeling perspective, removing these identifier-based features also helps prevent potential data leakage, ensuring that future models learn generalizable fraud patterns rather than memorizing transaction-specific identifiers.

## 10. Basic Dataset Summary

```python
total_transactions = len(df)
fraud_transactions = int(df[target_col].sum())
non_fraud_transactions = int((df[target_col] == 0).sum())
fraud_percentage = df[target_col].mean() * 100

summary_df = pd.DataFrame({
    "Metric": [
        "Total Transactions",
        "Fraudulent Transactions",
        "Legitimate Transactions",
        "Fraud Percentage"
    ],
    "Value": [
        total_transactions,
        fraud_transactions,
        non_fraud_transactions,
        round(fraud_percentage, 4)
    ]
})

summary_df
```

### Insight

The dataset contains a total of **1,852,394 transactions**, of which only **9,651 are fraudulent**, while **1,842,743 are legitimate**.

This results in a fraud rate of **0.5210%**, confirming that fraud is an extremely rare event in this dataset.

This severe class imbalance has important implications:

- A model that predicts all transactions as legitimate would still achieve over **99% accuracy**, making accuracy an unreliable metric
- Fraud detection models must prioritize **recall, precision, and F1-score** rather than accuracy
- Special techniques such as **class weighting, resampling (e.g., SMOTE), or anomaly detection approaches** may be required

From an analytical perspective, this imbalance also means that identifying patterns in fraudulent transactions requires careful comparison against a much larger base of legitimate behavior.

Overall, this confirms that fraud detection in this dataset is a **rare-event classification problem**, which significantly increases the complexity of building effective predictive models.

In practical terms, this means that successful fraud detection systems must focus on identifying a very small number of high-risk transactions within a vast majority of normal activity.

## 13. Log-Scaled Amount Analysis

Because transaction amount is heavily skewed, a log transformation can reveal the structure of the distribution more clearly.

```python
df["log_amt"] = np.log1p(df["amt"])

plt.figure(figsize=(10, 5))
sns.histplot(df["log_amt"], bins=60, kde=True)
plt.title("Log-Transformed Transaction Amount Distribution")
plt.xlabel("log(1 + Amount)")
plt.ylabel("Frequency")
plt.show()
```

### Insight

To address the strong right-skewness in transaction amounts, a logarithmic transformation (`log1p`) was applied to the `amt` variable.

The transformed distribution shows a **significant improvement in symmetry**, with values now more evenly spread across the range.

Key observations:
- the extreme high-value transactions are **compressed**, reducing their dominance
- the bulk of the data is now more **centrally distributed**, particularly around the mid-range values
- the distribution appears closer to **normal (bell-shaped)** compared to the original skewed distribution

This transformation is important because:
- it reduces the impact of outliers on analysis and modeling
- it allows patterns in smaller and medium transactions to become more visible
- many machine learning models perform better when features are more normally distributed

From a fraud detection perspective, this enables a more balanced comparison between low and high transaction values, ensuring that large transactions do not disproportionately influence the analysis.

Overall, the log transformation makes the transaction amount feature more suitable for both **statistical analysis and predictive modeling**.

The presence of multiple peaks in the transformed distribution may also suggest underlying behavioral segments, such as different spending patterns across transaction types or user groups.

## 17. Fraud Rate by Category

Fraud rate is more informative than raw fraud count because it adjusts for category size.

```python
category_summary = df.groupby("category").agg(
    total_transactions=(target_col, "count"),
    fraud_transactions=(target_col, "sum")
)
category_summary["fraud_rate_pct"] = 100 * category_summary["fraud_transactions"] / category_summary["total_transactions"]
category_summary = category_summary.sort_values("fraud_rate_pct", ascending=False)

category_summary
```

```python
plt.figure(figsize=(12, 6))
category_summary["fraud_rate_pct"].plot(kind="bar")
plt.title("Fraud Rate by Transaction Category (%)")
plt.xlabel("Category")
plt.ylabel("Fraud Rate (%)")
plt.xticks(rotation=45)
plt.show()
```

### Insight

While earlier analysis showed where fraud occurs most frequently, this section reveals where fraud is **most likely to occur**, providing a more accurate measure of risk.

The categories with the highest fraud rates are:
- **shopping_net:** 1.59%  
- **misc_net:** 1.30%  
- **grocery_pos:** 1.26%  

These results highlight a clear pattern:
- **online transaction categories (shopping_net, misc_net)** exhibit the highest fraud risk  
- this suggests that **card-not-present transactions** are more vulnerable due to reduced verification and increased exposure to digital threats  

Interestingly:
- **grocery_pos**, a physical transaction category, also shows a relatively high fraud rate  
- this may be due to **high-frequency environments**, where fraudulent transactions can blend into normal spending behavior  

A key observation is that:
- categories like **gas_transport** have the highest transaction volumes but relatively low fraud rates (~0.41%)  
- while **shopping_net**, with lower volume, has the **highest likelihood of fraud per transaction**  

This indicates that:
- **high transaction volume does not necessarily imply high fraud risk**  
- fraud is more strongly influenced by **transaction context than frequency**

Lower-risk categories include:
- **home, health_fitness, food_dining**, all with fraud rates around ~0.15%  

Overall, this analysis reveals that fraud risk is concentrated in:
- **online transactions**
- **specific behavioral contexts rather than just high-volume categories**

This reinforces the importance of incorporating **transaction category as a predictive feature** in fraud detection models, and highlights the value of using **fraud rates (not just counts)** for more accurate risk assessment.

- External factors (e.g., user behavior, system controls) are not captured

### 4. Static Snapshot
- Analysis is based on historical data
- Does not account for **evolving fraud strategies over time**

### 5. Geographic Features Underutilized
- Latitude/longitude were not deeply explored (e.g., clustering, distance anomalies)
- Potential spatial fraud patterns may remain hidden
